```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def barplot_biased_vs_unbiased(dataset="", biased=[], unbiased=[]):

    ################################################################################
    # Show correct/wrong shape/color bars
    ################################################################################
    
    plt.figure()
    
    # colors
    RED='#B71103'
    GREEN='#02854F'
    
    # bars
    plt.bar(x=0, height=biased[0], label='Correct prediction', width=0.4, color=GREEN)
    plt.bar(x=1, height=biased[1], label='Correct prediction', width=0.4, color=GREEN)
    plt.bar(x=2, height=biased[2], label='Wrong prediction', width=0.4, color=RED)
    plt.bar(x=3, height=biased[3], label='Wrong prediction', width=0.4, color=RED)
    
    # axes and labels
    plt.gca().set_xticklabels(['', 'Correct shape', '', 'Correct color', '', 'Wrong shape', '','Wrong color'])
    plt.ylabel('Probability in %')
    plt.title(f"Dominant concept of standalone model for {dataset} dataset")

    # make sure that each label does appear only once
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())

    plt.grid()
    plt.show()

    ################################################################################
    # Show color vs. shape
    ################################################################################
    
    plt.figure()
    plt.bar(x=[0.8, 1.8][0], height=biased[1] + biased[3], label='Color', width=0.2, color='#8c195fff')
    plt.bar(x=[1.2, 2.2][0], height=biased[0] + biased[2], label='Shape', width=0.2, color='#4E2A84ff')
    plt.gca().set_xticklabels(['', '', 'Color', '', '', '', 'Shape'])
    plt.ylabel('Probability in %')
    plt.title(f"Dominant concept of standalone model for {dataset} dataset")
    plt.legend()
    plt.grid()
    plt.show()
    
    ################################################################################
    # Show correct/wrong shape vs. any color (For unbiased grayscale model)
    ################################################################################
    
    plt.figure()
    
    # bars
    plt.bar(x=0, height=unbiased[0], label='Correct prediction', width=0.4, color=GREEN)
    plt.bar(x=1, height=unbiased[1], label='Wrong prediction', width=0.4, color=RED)
    plt.bar(x=2, height=unbiased[2], label='Wrong prediction', width=0.4, color=RED)
    
    # axes and labels
    plt.gca().set_xticklabels(['', 'Correct shape', '', 'Wrong shape', '', 'Any color', '', ''])
    plt.ylabel('Probability in %')
    plt.title(f"Dominant concept of unbiased standalone model for {dataset} dataset")

    # make sure that each label does appear only once
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())

    plt.grid()
    plt.show()

    ################################################################################
    # Show color vs. shape for biased and unbiased standalone model
    ################################################################################
    
    plt.figure()
    plt.bar(x=[1.2, 2.2], height=[biased[0] + biased[2], unbiased[0] + unbiased[1]], label='Shape', width=0.4, color='#4E2A84ff')
    plt.bar(x=[0.8, 1.8], height=[biased[1] + biased[3], unbiased[2]], label='Color', width=0.4, color='#8c195fff')
    plt.gca().set_xticklabels(['', '', 'Standalone model', '', '', '', 'Unbiased standalone model'])
    plt.ylabel('Probability in %')
    plt.title(f"Dominant concept for {dataset} dataset")
    plt.legend()
    plt.grid()
    plt.show()
    
```

```python
# training dataset (Data from biased and unbiased "Influence network surgery" column)
barplot_biased_vs_unbiased(dataset="training",
                           biased=[11.83, 38.75, 12.26, 37.16],
                           unbiased=[33.91, 32.06, 34.03])
```

```python
# test dataset (Data from biased and unbiased "Influence network surgery" column)
barplot_biased_vs_unbiased(dataset="test",
                           biased=[11.05, 38.90, 10.75, 39.30],
                           unbiased=[34.55, 31.10, 34.35])
```

```python
# real-world dataset (Data from biased and unbiased "Influence network surgery" column)
barplot_biased_vs_unbiased(dataset="real-world",
                           biased=[10.55, 41.10, 11.80, 36.55],
                           unbiased=[35.35, 34.20, 30.45])
```
