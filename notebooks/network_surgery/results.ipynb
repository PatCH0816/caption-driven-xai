```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def barplot_biased_vs_unbiased(title="", shape_vec=[], color_vec=[]):
    
    # shape_vec=[biased_shape_sum, unbiased_shape_sum]
    # color_vec=[biased_color_sum, unbiased_color_sum]
    
    plt.figure()
    # plt.bar(x=[1.2, 2.2], height=[21.8, 65.65], label='Shape', width=0.4, color='teal')
    # plt.bar(x=[0.8, 1.8], height=[78.2, 34.35], label='Color', width=0.4, color='salmon')
    # plt.bar(x=[1.2, 2.2], height=[21.8, 65.65], label='Shape', width=0.4, color='#8c195fff')
    # plt.bar(x=[0.8, 1.8], height=[78.2, 34.35], label='Color', width=0.4, color='#4E2A84ff')
    plt.bar(x=[1.2, 2.2][0], height=shape_vec[0], label='Shape', width=0.2, color='#4E2A84ff')
    plt.bar(x=[0.8, 1.8][0], height=color_vec[0], label='Color', width=0.2, color='#8c195fff')
    plt.gca().set_xticklabels(['', '', 'Color', '', '', '', 'Shape'])
    plt.ylabel('Probability in %')
    plt.title(title)
    # plt.legend()
    plt.grid()
    plt.show()
    
    plt.figure()
    # plt.bar(x=[1.2, 2.2], height=[21.8, 65.65], label='Shape', width=0.4, color='teal')
    # plt.bar(x=[0.8, 1.8], height=[78.2, 34.35], label='Color', width=0.4, color='salmon')
    # plt.bar(x=[1.2, 2.2], height=[21.8, 65.65], label='Shape', width=0.4, color='#8c195fff')
    # plt.bar(x=[0.8, 1.8], height=[78.2, 34.35], label='Color', width=0.4, color='#4E2A84ff')
    plt.bar(x=[1.2, 2.2], height=shape_vec, label='Shape', width=0.4, color='#4E2A84ff')
    plt.bar(x=[0.8, 1.8], height=color_vec, label='Color', width=0.4, color='#8c195fff')
    plt.gca().set_xticklabels(['', '', 'Biased standalone model', '', '', '', 'Unbiased standalone model'])
    plt.ylabel('Probability in %')
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
    
```
